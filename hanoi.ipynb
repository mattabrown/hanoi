```python
# tower of hanoi solver

!pip install -q mediapy

import mediapy as media
import numpy as np
import copy

def move_tower(state, size, start_pos, dest_pos, state_sequence):

    if (size==1):
        state[dest_pos].append(state[start_pos].pop())
        state_sequence.append(copy.deepcopy(state))
        return state,state_sequence

    other_pos = ({0,1,2}-{start_pos,dest_pos}).pop()

    state,state_sequence=move_tower(state,size-1,start_pos,other_pos,state_sequence)
    
    state[dest_pos].append(state[start_pos].pop())
    state_sequence.append(copy.deepcopy(state))
    
    state,state_sequence=move_tower(state,size-1,other_pos,dest_pos,state_sequence)

    return state,state_sequence

def draw_state_np(state,max_size,scale):    

    num_rows=max_size*scale
    num_cols=max_size*scale*3
    img=np.ones((num_rows,num_cols,3))
       
    for pos in range(3):    
        stack_count=0
        for sz in state[pos]:
            col0 = pos*max_size + max_size/2 - sz/2
            row0 = (max_size-1)-stack_count
            col1 = col0 + sz
            row1 = row0 + 1
            
            col0=int(col0*scale)
            row0=int(row0*scale)
            col1=int(col1*scale)
            row1=int(row1*scale)
            
            img[row0:row1,col0:col1,0:2]=0        
            stack_count=stack_count+1

    return img

def draw_state_sequence_np(state_sequence):
    
    max_size=10
    plot_scale=20
    num_states=len(state_sequence)
    
    img0=draw_state_np(state_sequence[0],max_size,plot_scale)
    rows,cols,bands=img0.shape
    imgs=np.zeros((num_states,rows,cols,bands))
    
    for i in range(num_states):
        imgs[i,:,:,:] = draw_state_np(state_sequence[i],max_size,plot_scale)
    
    return imgs
```

```python
# state=[[],[],[]]
state=[[9,8,7,6,5,4,3,2,1],[],[]]

state_sequence=[]
state_sequence.append(copy.deepcopy(state))
state,state_sequence=move_tower(state,9,0,1,state_sequence)

imgs=draw_state_sequence_np(state_sequence)

media.show_video(imgs,fps=8)
```
